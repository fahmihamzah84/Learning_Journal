## **Column Manipulation**
- rename column
- drop column
- create new column


```python
nba = df.copy()
nba.head()
```

```python
# rename nama kolom
nba.rename(columns={'game_location':'loc', 'game_result':'result'}, inplace=True)
```

```python
nba
```

```python
# Hapus kolom
nba.drop(labels=['elo_i', 'elo_n', 'opp_elo_i', 'opp_elo_n'], axis=1, inplace=True)
```

```python
nba
```

```python
nba['selisih'] = nba['pts'] - nba['opp_pts']
```

```python
nba
```

```python
nba.columns
```

**Latihan**

```python
# Siapkan list kosong
# definisikan argumen loop
# simpan hasil ke list kosong
# convert list menjadi kolom

hasil = []

for idx in range(len(nba)):
    selisih = nba.iloc[idx, 10] - nba.iloc[idx, 14]
    hasil.append(selisih)
hasil 
```

```python
nba['Selisihloop'] = hasil
```
